Словарь для перевода месяцев

```python
months = {"февраля": "02", "марта": "03"}
def clean_date(d):
    for k, v in months.items():
        d = d.replace(k, v)
    return d.replace(" г.", "")
```

```python
def analyze_vladivostok_vibes(data):
    cleaned_dates = [datetime.strptime(clean_date(d), "%d %m %Y") for d in data]
    df = pd.DataFrame({'date': cleaned_dates})
    
    df['weekday_num'] = df['date'].dt.weekday
    days_ru = ['пн', 'вт', 'ср', 'чт', 'пт', 'сб', 'вс']
    df['weekday'] = df['weekday_num'].apply(lambda x: days_ru[x])
    
    counts = df.groupby(['weekday_num', 'weekday']).size().reset_index(name='count')
    counts = counts.sort_values('weekday_num')
    total = counts['count'].sum()
    counts['share'] = counts['count'] / total
    
    min_share = counts['share'].min()
    counts['multiplier'] = (counts['share'] / min_share).round(2)
    
    print("Абсолютное количество комментариев по дням:\n", counts[['weekday', 'count']])
    print("\nДоли и множители активности:\n", counts[['weekday', 'share', 'multiplier']])
    
    plt.figure(figsize=(10, 6))
    bars = plt.bar(counts['weekday'], counts['count'], color='LightSkyBlue', edgecolor='Navy')
    
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{int(height)}', ha='center', va='bottom')
    
    plt.title('Активность по дням недели (VL.ru, февраль–март 2026)', fontsize=14)
    plt.xlabel('День недели')
    plt.ylabel('Количество комментариев')
    plt.grid(axis='y', linestyle='dashed', alpha=0.7)
    plt.tight_layout()
    plt.savefig('vlru_day_activity.png', dpi=150)
    plt.show()
    
    plt.figure(figsize=(10, 6))
    plt.bar(counts['weekday'], counts['share']*100, color='Salmon', edgecolor='darkred')
    plt.grid(axis='y', linestyle='dashed', alpha=0.4)
    plt.title('Доля активности по дням недели (%)')
    plt.xlabel('День недели')
    plt.ylabel('Процент от общего числа комментариев')
    for i, (_, row) in enumerate(counts.iterrows()):
        plt.text(i, row['share']*100 + 0.5, f"{row['share']*100:.1f}", ha='center')
    plt.tight_layout()
    plt.savefig('vlru_day_share.png', dpi=150)
    plt.show()

analyze_vladivostok_vibes(raw_data)
```
